```python
import pandas as pd

df = pd.read_csv('sales-data.csv')
df.head(3)
```

According to the author:

 - ad_id: an unique ID for each ad.

 - xyzcampaignid: an ID associated with each ad campaign of XYZ company.

 - fbcampaignid: an ID associated with how Facebook tracks each campaign.

 - age: age of the person to whom the ad is shown.

 - gender: gender of the person to whim the add is shown

 - interest: a code specifying the category to which the person’s interest belongs (interests are as mentioned in the person’s Facebook public profile).

 - Impressions: the number of times the ad was shown.

 - Clicks: number of clicks on for that ad.

 - Spent: Amount paid by company xyz to Facebook, to show that ad.

 - Total conversion: Total number of people who enquired about the product after seeing the ad.

 - Approved conversion: Total number of people who bought the product after seeing the ad.

Let's remove the underscores from the column names and capitalize the first letter of each word of the column name, then change some of the column names to remove confusion.

```python
df = df.rename(columns={cname: " ".join([word.capitalize() for word in cname.split('_')]) for cname in df.columns})
df = df.rename(columns={'Ad Id': 'Ad ID', 'Xyz Campaign Id': 'Company Campaign ID', 
                        'Fb Campaign Id': 'FB Campaign ID', 'Total Conversion': 'Inquiries', 'Approved Conversion': 'Sales'})
```

Let's plot the distributions of the numerical columns to find if some are actually categorical.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')

fig, axs = plt.subplots(3, 3)
fig.set_size_inches(20, 18)

spent = sns.histplot(x='Spent', data=df, ax=axs[0][0])
spent.set_title('Frequency of the ad cost')

impressions = sns.histplot(x='Impressions', data=df, ax=axs[0][1])
impressions.set_title('Frequency of impressions')

clicks = sns.histplot(x='Clicks', data=df, ax=axs[0][2])
clicks.set_title('Frequency of clicks')

inquiries = sns.histplot(x='Inquiries', data=df, ax=axs[1][0])
inquiries.set_title('Frequency of inquiries')

sales = sns.histplot(x='Sales', data=df, ax=axs[1][1])
sales.set_title('Frequency of sales')

company_campaign_ids = sns.histplot(x='Company Campaign ID', data=df, ax=axs[1][2])
company_campaign_ids.set_title('Frequency of campaign IDs')

interest_count = sns.histplot(x='Interest', data=df, ax=axs[2][0])
interest_count.set_title('Frequency of interests')

ad_id = sns.histplot(x='Ad ID', data=df, ax=axs[2][1])
ad_id.set_title('Frequency of Ad IDs')

fb_campaign_id = sns.histplot(x='FB Campaign ID', data=df, ax=axs[2][2])
fb_campaign_id.set_title('Frequency of Facebook Ad IDs')
```

Most of the distributions appear to be zero centered except for:
 - Company Campaign ID's
 - Interests
 - Ad ID's
 - Facebook Ad ID's
 
which due to the shapes of their distributions fitting into defined categories appear to be categorical. Let's investigate this.

To do this let's take the number of unique values in the main dataframe 'df' and the unique values for each column and make it into it's own seperate dataframe to determine if the data is categorical or not.

```python
categorical_hypothesis_cols = ['Company Campaign ID', 'Interest', 'Ad ID', 'FB Campaign ID']
categorical_hypothesis_dic = {'Column Name': categorical_hypothesis_cols, 
       'Number Of Unique Values': [df[cname].nunique() for cname in categorical_hypothesis_cols], 
       'Unique Values': [df[cname].unique() for cname in categorical_hypothesis_cols]}
categorical_hypothesis_df = pd.DataFrame(categorical_hypothesis_dic)

categorical_hypothesis_df
```

It can be seen that there are 3 unique 'Company Campaign ID' values despite there being 1143 values given, meaning that it is likely to be categorical. This goes for the 'Interest' column too, as there are only 40 unique interests, despite there being 1143 values given. This means that it is most likely categorical too. The 'FB Campaign ID' column is most likely non-categorical based on the number of unique values, most likely occuring due to reused ID's, as it was stated in the description of the column that the Facebook Campaign ID's were not unique and was just how Facebook tracks a campaign, and therefore having no real significance. All of the values are unique for the 'Ad ID' column, therefore resulting in it being non-categorical.

### Analysis

#### Who are clicking on the ads?

Let's start off by comparing the clicks vs impressions between each gender.

```python
clicks_impressions = sns.lmplot(x='Impressions', y='Clicks', hue='Gender', data=df, height=6, aspect=1.5)
clicks_impressions.fig.suptitle('Clicks vs impressions')
```

It can be seen that females are more likely to click on an ad than males. Let's investigate that age these females are and what interests they have.

```python
female_df = df[df['Gender'] == 'F']
clicks_impressions = clicks_impressions = sns.lmplot(x='Impressions', y='Clicks', hue='Age', data=female_df, height=6, aspect=1.5)
clicks_impressions.fig.suptitle('Clicks vs impressions for females')
```

So it can be seen that females aged 45-49 had the highest amount of clicks. Let's find the interest with the highest clickrate for this demographic of 45-49 year old females.

Let's find the rows that were targetted at 45-49 year old females and create a click rate column.

```python
female_age_df = female_df[female_df['Age'] == '45-49']
female_age_df['Click Rate'] = female_age_df['Clicks'] / female_age_df['Impressions']
female_age_df['Click Rate'] = female_age_df['Click Rate'].apply(lambda x: 0 if np.isinf(x) or np.isnan(x) else x)
```

Let's also sort this dataframe by the number at the end of the interests in the 'Interest' column.

```python
female_age_df['Temp Col'] = female_age_df['Interest'].apply(lambda x: int(x.split('_')[1]))
female_age_df = female_age_df.sort_values(by='Temp Col')
female_age_df = female_age_df.drop('Temp Col', axis=1)
female_age_df = female_age_df.reset_index()
```

Now let's plot this click rate against the interest to see what the interest shared by 45-49 year old females that had the highest click rate.

```python
fig, axs = plt.subplots(figsize=(16, 6))

click_rate_interest = sns.barplot(x='Interest', y='Click Rate', data=female_age_df, ci=None)
click_rate_interest.set_xticklabels(labels=click_rate_interest.get_xticklabels(), rotation=90)
click_rate_interest.set_title('Click rate vs interests for 45-49 year old females')
```

It can be seen that the interest shared among 45-49 year old women with the highest clickrate was 'Interest_10'. In summary these results demonstrate that in order to improve the clicks for their advertisements, advertisers should specifically target females that are 45-49 years old and have an interest of 'Interest_10'.

#### Who are most likely to make a sale?

Let's start off by comparing the sales vs clicks for each gender.

```python
sales_clicks = sns.lmplot(x='Clicks', y='Sales', hue='Gender', data=df, height=6, aspect=1.5)
sales_clicks.fig.suptitle('Sales vs clicks')
```

This demonstrates to us that males are more likely to make a purchase after clicking on an ad. Let's further investigate the age of the males that are making these purchases.

```python
male_df = df[df['Gender'] == 'M']
sales_clicks = sns.lmplot(x='Clicks', y='Sales', hue='Age', data=male_df, height=6, aspect=2.5)
sales_clicks.fig.suptitle('Sales vs clicks for males')
```

It can be seen that 30-34 year old males had the highest amount of sales compared to their clicks. Let's find the interest with the sales per click (conversion rate) for this demographic of 30-34 year old males.

Let's find the rows that were targetted at 30-34 year old males and create a conversion rate column.

```python
male_age_df = male_df[male_df['Age'] == '30-34']
male_age_df['Conversion Rate'] = male_age_df['Sales'] / male_age_df['Clicks']
male_age_df['Conversion Rate'] = male_age_df['Conversion Rate'].apply(lambda x: 0 if np.isinf(x) or np.isnan(x) else x)
```

Let's also sort this dataframe by the number at the end of the interests in the 'Interest' column.

```python
male_age_df['Temp Col'] = male_age_df['Interest'].apply(lambda x: int(x.split('_')[1]))
male_age_df = male_age_df.sort_values(by='Temp Col')
male_age_df = male_age_df.drop('Temp Col', axis=1)
male_age_df = male_age_df.reset_index()
```

Now let's plot this click rate against the interest to see what the interest shared by 30-34 year old males that had the conversion rate.

```python
fig, axs = plt.subplots(figsize=(16, 6))

conversion_rate_interest = sns.barplot(x='Interest', y='Conversion Rate', data=male_age_df, ci=None)
conversion_rate_interest.set_xticklabels(labels=conversion_rate_interest.get_xticklabels(), rotation=90)
conversion_rate_interest.set_title('Conversion rate vs interests for 30-34 year old males')
```
